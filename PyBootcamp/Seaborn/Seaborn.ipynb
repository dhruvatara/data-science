```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
#loading a dataset from seaborn
tips = sns.load_dataset('tips')
tips.head()
```

<b> Disribution plot </b>

```python
#making the distplot
sns.distplot(tips['total_bill'])
```

```python
#removing kde and defining bins
sns.distplot(tips['total_bill'],bins=10,kde=False)
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
#can also change type of plot with kind=reg, hex or kde
```

```python
#making a pairplot
sns.pairplot(tips,hue='smoker',palette='coolwarm')
#can change hue to another categorical variable or palette to another seaborn palette
```

```python
#making a rugplot, which forms the basis of a kde plot
sns.rugplot(tips['total_bill'])
```

<p> KDE stands for kernel density estimation. Click <a href="https://en.wikipedia.org/wiki/Kernel_density_estimation">here</a> for more.

<b>Categorical Plots</b>

```python
#making a barplot
sns.barplot(x='sex',y='total_bill',data=tips,hue='smoker',estimator=np.std)
#can use hue to show another categorical value, can sue estimator to change default y axis from mean to other aggregator
```

```python
#making a countplot
sns.countplot(x='sex',data=tips,hue='day')
#countplot explicitly only counts the occurence of the categorical data on the x axis
```

```python
#making a boxplot
sns.boxplot(x='sex',y='total_bill',data=tips,hue='smoker',palette='coolwarm')
```

```python
#making a violinplot
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True,palette='coolwarm')
```

```python
#making a stripplot
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
#making a factorplot
sns.catplot(x='day',y='tip',data=tips,kind='ba')
```
