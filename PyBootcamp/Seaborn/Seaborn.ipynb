```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
#loading a dataset from seaborn
tips = sns.load_dataset('tips')
tips.head()
```

<b> Disribution plot </b>

```python
#making the distplot
sns.distplot(tips['total_bill'])
```

```python
#removing kde and defining bins
sns.distplot(tips['total_bill'],bins=10,kde=False)
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
#can also change type of plot with kind=reg, hex or kde
```

```python
#making a pairplot
sns.pairplot(tips,hue='smoker',palette='coolwarm')
#can change hue to another categorical variable or palette to another seaborn palette
```

```python
#making a rugplot, which forms the basis of a kde plot
sns.rugplot(tips['total_bill'])
```

<p> KDE stands for kernel density estimation. Click <a href="https://en.wikipedia.org/wiki/Kernel_density_estimation">here</a> for more.

<b>Categorical Plots</b>

```python
#making a barplot
sns.barplot(x='sex',y='total_bill',data=tips,hue='smoker',estimator=np.std)
#can use hue to show another categorical value, can sue estimator to change default y axis from mean to other aggregator
```

```python
#making a countplot
sns.countplot(x='sex',data=tips,hue='day')
#countplot explicitly only counts the occurence of the categorical data on the x axis
```

```python
#making a boxplot
sns.boxplot(x='sex',y='total_bill',data=tips,hue='smoker',palette='coolwarm')
```

```python
#making a violinplot
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True,palette='coolwarm')
```

```python
#making a stripplot
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
#making a factorplot
sns.catplot(x='day',y='tip',data=tips,kind='violin')
```

<b>Matrix Plots</b>

```python
#making your data into matrix
tc = tips.corr()
flights = sns.load_dataset('flights')
flights.head()
fc = flights.pivot_table(index = 'month',columns='year',values='passengers')
```

```python
#making a heatmap
sns.heatmap(tc,cmap='magma',linewidth='1',linecolor='pink')
#use cmap to specify color scale, linecolor and linewidth to make a distinction b/w diff regions
```

```python
#making a clustermap
sns.clustermap(fc,cmap='magma',standard_scale=1)
```

<b> Regression Plot </b>

```python
#making one using lmplot
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','+'],scatter_kws={'s':50},col='day')
#scatter_kws is used to make changes directly to the matplotlib that seaborn calls to make the plot
#col and row can be used to introduce grids of plots
#we can change the size of plots using aspect and size params
```
