<b> Basics of Plotting a Graph

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

<ul>
    <li>Functional method</li>
</ul>    

```python
#making data points
x = np.arange(0,6)
y = x**2
```

```python
#plotting basic graph
plt.plot(x,y)
```

```python
#adding x and y labels
plt.xlabel('Number')
plt.ylabel('Square')
plt.plot(x,y,'b--')
```

```python
#multiplot 2 graphs on same canvas
plt.subplot(1,2,1)
plt.plot(x,y,'b')
plt.subplot(1,2,2)
plt.plot(y,x,'r')
```

<ul>
    <li>OO method</li>
</ul>

```python
#making a figure
fig = plt.figure()
#adding axes
ax = fig.add_axes([0.0,0.0,0.6,0.6])
ax.set_xlabel('Number')
ax.set_ylabel('Squares')
ax.set_title('Square graph')
ax.plot(x,y)
```

```python
#multiplot 2 canvas in one
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0,0,0.3,0.3])
ax.plot(x,y,'r')
ax2.plot(y,x)
```

```python
#indexing axes
fig, ax = plt.subplots(nrows=2,ncols=1)
ax[0].plot(x,y)
ax[1].plot(y,x,'r')
plt.tight_layout()
```

<b> Firgure size and DPI

```python
#setting figsize and DPI
fig = plt.figure(figsize=(10,5),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_ylabel('Sq.')
ax.set_xlabel('No.')
```

```python
#saving graph to file
fig.savefig('gr.jpg',dpi=300)
```

```python
#displaying and moving legend
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label='X squared')
ax.plot(x,x**3,'r',label='X cubed')
ax.set_ylabel('Sq.')
ax.set_xlabel('No.')
ax.legend(loc=10)
```
